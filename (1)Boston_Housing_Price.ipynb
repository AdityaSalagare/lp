### Problem Statement

**Linear regression by using Deep Neural network**: Implement Boston housing price prediction problem by Linear regression using Deep Neural network. Use Boston House price prediction dataset.

### Import Library

```python
# Data analysis and visualization
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Preprocessing and evaluation
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler
```

### Load Data

```python
(X_train , y_train), (X_test , y_test) = tf.keras.datasets.boston_housing.load_data(
                                            path = 'boston_housing_npz',
                                            test_split = 0.2,
                                            seed = 42
                                        )
```

```python
# Converting Data to DataFrame
X_train_df = pd.DataFrame(X_train)
y_train_df = pd.DataFrame(y_train)

# Preview the training data
X_train_df.head(10)
```

## Preprocessing

```python
# Create column transformer
ct = make_column_transformer(
    (MinMaxScaler(), [0,1,2,4,5,6,7,8,9,10,11,12]))

X_train = ct.fit_transform(X_train).astype('float32')
X_test  = ct.transform(X_test).astype('float32')
y_train = y_train.astype('float32')
y_test  = y_test.astype('float32')

# Distribution of X_train feature values after normalization
pd.DataFrame(X_train).describe()
```

# Model, Predict, Evaluation

```python
# Reserve data for validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

## Creating the Model and Optimizing the Learning Rate

learning rate = 0.01,
batch_size = 32,
dense_layers = 2,
hidden_units for Dense_1 layer= 10,
hidden_units for Dense_2 layer = 100

```python
# Building the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu',
                          kernel_regularizer=tf.keras.regularizers.l2(1e-3),
                          input_shape=(X_train.shape[1],),
                          name='Dense_1'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32, activation='relu',
                          kernel_regularizer=tf.keras.regularizers.l2(1e-3),
                          name='Dense_2'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, name='Prediction')
])

# Compiling the model
model.compile(
    loss='mse',
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    metrics=['mse']
)

# Training the model
history = model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=200,
    validation_data=(X_val, y_val),
    verbose=2
)

# Evaluating the model
test_loss, test_mse = model.evaluate(X_val, y_val)
print(f"Test Loss: {test_loss:.4f}, Test MSE: {test_mse:.4f}")

# 6) Evaluate on real test set
test_loss1, test_mse1 = model.evaluate(X_test, y_test, verbose=0)
test_rmse1 = np.sqrt(test_mse)
print(f"Test MSE:  {test_mse1:.4f}")
print(f"Test RMSE: {test_rmse1:.4f} thousand USD")

# Making predictions
predictions = model.predict(X_test[:5])
print("Sample predictions vs true:",
      list(zip(predictions.flatten(), y_test[:5])))
```
